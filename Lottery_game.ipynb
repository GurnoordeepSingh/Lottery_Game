## Python Lottery Draw Game

Welcome to our simple text-based lottery draw game! Here's how it works:

*   There will be 5 players.
*   Each player starts with 100 virtual coins.
*   The entry fee for a game can vary (starting from 10 coins up to 100 coins).
*   The system will generate a set of lottery tickets, including winning tickets.
*   Players will choose one ticket from the available set.
*   The system will lock the chosen ticket to each player.
*   The winning tickets will be revealed, and the prize pool will be distributed among the winners based on their chosen ticket:
    *   1st place: 50% of the prize pool
    *   2nd place: 75% of the remaining prize pool
    *   3rd place: The rest of the remaining prize pool
    *   4th and 5th place: No winnings (if they chose a losing ticket)
*   After each round, player coin balances will be updated, and the game can be reset for a new round with new tickets.

Let's get started with setting up the players!

```python
# Initialize players with their names and starting coins
players = {
    "player1": {"coins": 100, "chosen_ticket": None},
    "player2": {"coins": 100, "chosen_ticket": None},
    "player3": {"coins": 100, "chosen_ticket": None},
    "player4": {"coins": 100, "chosen_ticket": None},
    "player5": {"coins": 100, "chosen_ticket": None}
}

print("Players initialized:")
for player, data in players.items():
    print(f"- {player}: {data['coins']} coins")
```

```python
import random

# This cell is for initial setup and can remain outside the main game loop.
# We'll generate available ticket numbers for the very first round here.
num_tickets = len(players)
available_ticket_numbers = list(range(1, num_tickets + 1))

print(f"Ticket numbers from 1 to {num_tickets} will be available for players to choose in the first round.")
```

```python
2
# This cell will now be part of the main game loop.
# It handles getting player choices and determining winning tickets from those choices.

# Present the available tickets to the players
print("\nAvailable tickets:")
print(available_ticket_numbers) # Use the available_ticket_numbers generated before the loop or reset in the loop

# Get ticket choices from each player
print("\nPlayers, please choose a ticket from the available numbers:")

# Reset chosen tickets for the new round
for player in players:
    players[player]["chosen_ticket"] = None

# Make a copy of available tickets for this round to remove from as players choose
current_available_tickets = available_ticket_numbers[:]

for player in players:
    while True:
        try:
            # Ensure we are only showing and allowing selection from currently available tickets
            print(f"Available tickets for {player}: {current_available_tickets}")
            pick = int(input(f"{player}, enter your chosen ticket number: "))
            if pick in current_available_tickets:
                players[player]["chosen_ticket"] = pick
                current_available_tickets.remove(pick) # Remove the chosen ticket so others can't pick it
                print(f"{player} has chosen ticket number {pick}")
                break
            else:
                print("Invalid ticket number. Please choose an available ticket.")
        except ValueError:
            print("Invalid input. Please enter a valid ticket number.")

print("\nAll players have selected their tickets.")

# Now that players have chosen their tickets, we select the winning tickets from those chosen.
chosen_tickets_list = [data["chosen_ticket"] for data in players.values()]
random.shuffle(chosen_tickets_list) # Shuffle the chosen tickets to randomly pick winners

# Assign winning tickets from the shuffled list of chosen tickets
# We need exactly 3 winning tickets for 1st, 2nd, and 3rd place from the chosen tickets
if len(chosen_tickets_list) < 3:
     print("Error: Not enough chosen tickets to assign 1st, 2nd, and 3rd prizes.")
     winning_tickets = {} # No prizes if not enough chosen tickets
else:
    winning_tickets = {
        "1st_prize": chosen_tickets_list.pop(0), # Remove and assign the first chosen ticket as 1st prize
        "2nd_prize": chosen_tickets_list.pop(0), # Remove and assign the next chosen ticket as 2nd prize
        "3rd_prize": chosen_tickets_list.pop(0)  # Remove and assign the next chosen ticket as 3rd prize
    }
    # The remaining tickets in chosen_tickets_list were not winning tickets

print(f"\nWinning tickets have been selected from the tickets chosen by the players.")
# We will reveal the winning tickets and distribute prizes in the next step.
```

```python
# This cell will now be part of the main game loop.
# It handles revealing winning tickets, determining winners, and distributing prizes.

# Reveal the winning tickets
print("\n--- Revealing Winning Tickets ---")
if winning_tickets:
    for prize, ticket in winning_tickets.items():
        print(f"{prize}: Ticket number {ticket}")
else:
    print("No winning tickets this round due to not enough chosen tickets.")


# Determine which players chose winning tickets and their prize level
prize_winners = {
    "1st_prize": [],
    "2nd_prize": [],
    "3rd_prize": []
}

if winning_tickets: # Only proceed if there are winning tickets
    for player, data in players.items():
        chosen_ticket = data["chosen_ticket"]
        if chosen_ticket == winning_tickets.get("1st_prize"):
            prize_winners["1st_prize"].append(player)
        elif chosen_ticket == winning_tickets.get("2nd_prize"):
            prize_winners["2nd_prize"].append(player)
        elif chosen_ticket == winning_tickets.get("3rd_prize"):
            prize_winners["3rd_prize"].append(player)

# Calculate and distribute the prize pool
# We'll use a fixed entry fee for now as in the previous version
entry_fee = 10
prize_pool = entry_fee * len(players)

print(f"\nEntry fee per player: {entry_fee}")
print(f"Total prize pool: {prize_pool} coins")

# Deduct entry fee
for player in players:
    players[player]["coins"] -= entry_fee


print("\n--- Prize Distribution ---")

if winning_tickets: # Only distribute prizes if there are winning tickets
    # 1st Place Prize
    first_place_prize = prize_pool * 0.50
    if prize_winners["1st_prize"]:
        prize_per_winner = first_place_prize / len(prize_winners["1st_prize"])
        for winner in prize_winners["1st_prize"]:
            players[winner]["coins"] += prize_per_winner
            print(f"1st Place Winner: {winner} wins {prize_per_winner:.2f} coins!")
    else:
        print("No 1st place winner this round.")

    remaining_prize_pool_after_1st = prize_pool - first_place_prize

    # 2nd Place Prize
    second_place_prize = remaining_prize_pool_after_1st * 0.75
    if prize_winners["2nd_prize"]:
        prize_per_winner = second_place_prize / len(prize_winners["2nd_prize"])
        for winner in prize_winners["2nd_prize"]:
            players[winner]["coins"] += prize_per_winner
            print(f"2nd Place Winner: {winner} wins {prize_per_winner:.2f} coins!")
    else:
        print("No 2nd place winner this round.")

    remaining_prize_pool_after_2nd = remaining_prize_pool_after_1st - second_place_prize

    # 3rd Place Prize
    third_place_prize = remaining_prize_pool_after_2nd
    if prize_winners["3rd_prize"]:
        prize_per_winner = third_place_prize / len(prize_winners["3rd_prize"])
        for winner in prize_winners["3rd_prize"]:
            players[winner]["coins"] += prize_per_winner
            print(f"3rd Place Winner: {winner} wins {prize_per_winner:.2f} coins!")
    else:
        print("No 3rd place winner this round.")

    # Players who did not win
    for player in players:
        if player not in prize_winners["1st_prize"] and player not in prize_winners["2nd_prize"] and player not in prize_winners["3rd_prize"]:
            print(f"{player} did not win this round.")
else:
    print("\nPrize distribution skipped due to not enough chosen tickets.")
```
